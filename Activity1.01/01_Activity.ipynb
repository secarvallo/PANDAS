# Data visualization

```python
# Visualiza el numero de estudiantes por cada categoria de tamano de familia
aggData = studentData.groupby(['famsize'])['famsize'].\
agg('count')
aggData
```

```python
# Para graficar esta data, tu necesitas crear los valores de x y la y-axi. Para definir estos, con la ayuda de indeces unicos del grupo de data, you puedes agregar los siguientes codigos. 
x = list(aggData.index)
x
```

```python
y = aggData.values
y
```

```python
# Importa la libreria matplot
import matplotlib.pyplot as plt
# Se utiliza el estilo de grafico called ggplot
plt.style.use('ggplot')

# EL plt.bar() metodo plots de la data se encuentra el tipo de grafico de barra. Dentro de este metodo, tu puedes definir los valores x-axis y la y-axis y tambien define el color.
plt.bar(x, y, color='gray')
plt.xlabel('Family Sizes')
plt.ylabel('Count of Students')
plt.title('Distributions of students against family sizes')
plt.show()
```

# Time series data

```python
# Covierte un string dentro de date object
date = pd.to_datetime('15th of January, 2021')
print(date)
```

```python
# Fecha despues de 25 días
newdate = date+pd.to_timedelta(25,unit='D')
# El parametro unit='D' es usada par definir 
print(newdate)
```

```python
# Obtine todos los datos dentro lo siguientes 7 días.
futureDate = pd.date_range(start=newdate, periods=7, freq='D') # freq = 'D' indica que la unida es el Día.
# En el siguiente ejercicio, La funcion date_range() fue utilizada para obtener una lista de todo las fechas futuras.
futureDate
```

# Code optimization

```python
# La funcion por cada elemento de las caracteristicas de Grado esta incrementa en 5 
def add5 (x):
    return x + 3
```

```python
# Utilizando el metodo creado
df = studentData[['G1', 'G2', 'G3']].apply(add5)
df
```

```python
# Esto puede ser logrado sin definir una funcion separada utilizando el metodo lambda
# El metodo Lambda te permite ejecutar la operation en justo solo una linea de codigo. 
df = studentData[['G1', 'G2', 'G3']].apply(lambda x:x+5)
df.head()
```
