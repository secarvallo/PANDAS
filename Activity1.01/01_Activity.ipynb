# The Series Object

La data unidimensional es representada por objetos de series.


```python
# Creating a Series
# Con el siguiente codigo muestra el constructor que es usado para crear una nueva serie.
ser1 = pd.Series([10,20,30,40])
# Displaying the Series
ser1
```

```python
# Se pueden representar diferentes tipos de datos en las Series
ser2 =pd.Series([[10, 20],\
                 [30, 40.5, 'series'],\
                 [50, 55],\
                 {'Name': 'Tess', 'Org':'Packt'}])
ser2
```

# The Dataframe object

Un Dataframe es una representacion bi-dimensional de data en filas y columnas,


```python
# Crear un Dataframe using the constructor
df = pd.DataFrame([30,50,20])
# En el codigo anterior, una lista de tres elementos fue convertido dentro del Constructor DataFrame()
# Mostrar
df
```

```python
# La salida es un DataFrame que puede ser visualizado usando el df.shape()
df.shape
```

```python
'''
Tu podrias ver 0 en la parte superior de la columna. Este es el nombre po defecto que este podria ser asignado a la columna cuando el Dataframe es creado.
Tu tambien podrias ver el numero 0, 1  y 2 a lo largo de las filas. Estas son llamadas indices.
'''
# Para mostrar las nombre de las columnas de los DataFrame, tu puedes usar el siguiente comando df.columns
df.columns
```

```python
df.index
```

```python
# Tu puedes convertir estos dentro de una lista utilizando usando el siguiente comando list(). EL sioguiente atajo convierte los nombres de las columnas y los indicies de las filas dentro de listas y estos imprimen sus valoes.
print('These are the names od the columns', list(df.columns))
print('These are the row indices', list(df.index))
# Puedes ver esto en los nombres de las columnas y los indices de las filas representadas en una lista.
```

```python
# Tambien puedes renombra las columnas y los indices de las filas para asignar estos para cualquier lista de valor.
# Este comando para renombrar una columna es df.columns.
# REnombra las columnas 
df.columns = ['V1']
df
```

```python
# Renombrar los indices
df.index = ['R1', 'R2', 'R3']
df
```

```python
# Creacion de DataFrame con multiple columnas 
df1 = pd.DataFrame([[10, 15, 20],\
                    [100, 200, 300]])
print(f'Shape of new data frame', df1.shape)
df1
```

```python
# Se pueden asignar los nombre de las columnas y los nombres de filas mientras se crea el DataFrame
df1 = pd.DataFrame([[10,15,20],[100,200,300]],\
                    columns= ['V1', 'V2', 'V3'],\
                    index= ['R1', 'R2'])
df1
```

# Working with local files

```python
# EL prim,er paso es importar la data es para definir la ruta para el CSV, el segundo parametro es el delimitador, especifica como las diferentes columnas de data son separado dentro de el archivo.
studentData = pd.read_csv('../Activity1.01/student/student-por.csv', delimiter=';')
```

```python
studentData.head(5)
```

```python
# Se pueden obtener snapshot de los dataset para visualizar solo las primeras y ultimas pocas filas.  Este es donde el head() y tail() las funciones resultan utiles.
studentData.tail(5)
```

```python
# Toma las primeras cinco filas de la data este se almacena en otra variable llamada studentSmall.
studentSmall = studentData.head()
# Se define la ruta de salida para escribir el archivo
outpath = '../Activity1.01/student/studentSmall.csv'
```

```python
# Escribe la data en el directorio en el disco 
studentSmall.to_csv(outpath)
```
